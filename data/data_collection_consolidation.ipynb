```
import pandas as pd
import numpy as np
from pprint import pprint
from bs4 import BeautifulSoup
import requests
```

```
TPGS = []
for i in range(1979, 2020):
  if i not in list(range(1999,2009)):
    TPGS.append(str(i)+'-'+str(int(str(i)[-2:]) + 1)+'TPGS.csv')
  elif i == 1999:
    TPGS.append(str(i)+'-00'+'TPGS.csv')
  else:
    TPGS.append(str(i)+'-0'+str(int(str(i)[-2:]) + 1)+'TPGS.csv')
OPGS = []
for i in range(1979, 2020):
  if i not in list(range(1999,2009)):
    OPGS.append(str(i)+'-'+str(int(str(i)[-2:]) + 1)+'OPGS.csv')
  elif i == 1999:
    OPGS.append(str(i)+'-00'+'OPGS.csv')
  else:
    OPGS.append(str(i)+'-0'+str(int(str(i)[-2:]) + 1)+'OPGS.csv')
```

```
champion = ['Los Angeles Lakers','Toronto Raptors','Golden State Warriors','Golden State Warriors','Cleveland Cavaliers','Golden State Warriors',
            'San Antonio Spurs','Miami Heat','Miami Heat','Dallas Mavericks','Los Angeles Lakers','Los Angeles Lakers','Boston Celtics',
            'San Antonio Spurs','Miami Heat','San Antonio Spurs','Detroit Pistons','San Antonio Spurs','Los Angeles Lakers','Los Angeles Lakers',
            'Los Angeles Lakers','San Antonio Spurs','Chicago Bulls','Chicago Bulls','Chicago Bulls','Houston Rockets','Houston Rockets','Chicago Bulls','Chicago Bulls',
            'Chicago Bulls','Detroit Pistons','Detroit Pistons','Los Angeles Lakers','Los Angeles Lakers','Boston Celtics','Los Angeles Lakers',
            'Boston Celtics','Philadelphia 76ers','Los Angeles Lakers','Boston Celtics','Los Angeles Lakers']
season = []
for i in range(1979, 2020):
  if i not in list(range(1999,2009)):
    season.append(str(i)+'-'+str(int(str(i)[-2:]) + 1))
  elif i == 1999:
    season.append(str(i)+'-00')
  else:
    season.append(str(i)+'-0'+str(int(str(i)[-2:]) + 1))
```

```
def season_combiner(csv_file1, csv_file2, champion, season): # pass in team stats per game, opponent stats per game, and season champion in that order
  dataframe1 = pd.read_csv(csv_file1)
  dataframe2 = pd.read_csv(csv_file2)
  dataframe1['Team'] = dataframe1['Team'].str.rstrip('*')
  dataframe2['Team'] = dataframe2['Team'].str.rstrip('*')
  season_toReturn = dataframe1.merge(dataframe2, on = 'Team', suffixes = ('_team', '_opponent'))
  season_toReturn['champion'] = np.where(season_toReturn['Team'] == champion, 1, 0)
  season_toReturn['season'] = season
  return season_toReturn
```

```
nba_seasons = pd.DataFrame()
for i in range(len(TPGS)):
  nba_seasons = pd.concat([nba_seasons, season_combiner(TPGS[i],OPGS[i],champion[i],season[i])])
```
